```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.amazon.in/s?k=phone&crid=AOZ5NTXZ736C&sprefix=phone%2Caps%2C309&ref=nb_sb_noss_2"
```

```python
# Headers for request
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36 Edg/134.0.0.0', 'Accept-Language': 'en-US, en;q=0.5'}) #add your user agent 
```

```python
# HTTP Request
webpage = requests.get(URL, headers=HEADERS)
```

```python
webpage
```

```python
type(webpage.content)
```

```python
# Soup Object containiang all data
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

```python
# Fetch links as List of Tag Objects
links = soup.find_all("a", attrs={'class':"a-link-normal s-line-clamp-2 s-link-style a-text-normal"})
```

```python
links
```

```python
link = links[0].get('href')
```

```python
link
```

```python
product_list = "https://amazon.in" + link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
```

```python
new_webpage
```

```python
# Soup Object containiang all data
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
```

```python
def get_category(soup):
    try:
        table = soup.find("table", attrs={"id": "productDetails_detailBullets_sections1"})
        if table:
            rows = table.find_all("tr")
            for row in rows:
                header = row.find("th")
                value = row.find("td")
                if header and value:
                    if "Generic Name" in header.text.strip():
                        return value.text.strip()
    except AttributeError:
        return "Not Available"
    
    return "Not Available"
```

```python
get_category(new_soup)
```
